# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import pickle
import json
import warnings

warnings.filterwarnings("ignore")
```

# 2. Load Data Files

```python
with open('list_num_cols.txt', 'r') as file_1:
    list_num_cols = json.load(file_1)

with open('list_cat_cols_nom.txt', 'r') as file_2:
    list_cat_cols_nom = json.load(file_2)

with open('list_cat_cols_ord.txt', 'r') as file_3:
    list_cat_cols_ord = json.load(file_3)

with open('model_scaler.pkl', 'rb') as file_4:
    scaler = pickle.load(file_4)

with open('model_encoder.pkl', 'rb') as file_5:
    encoder = pickle.load(file_5)

with open('model_ohe.pkl', 'rb') as file_6:
    ohe = pickle.load(file_6)

with open('model_lin_reg.pkl', 'rb') as file_7:
    model = pickle.load(file_7)
```

# 3. Retrieve Dataset for Inference Purposes

```python
with open('X_inference.txt', 'r') as file_8:
    X_inference = json.load(file_8)

with open('y_inference.txt', 'r') as file_9:
    y_inference = json.load(file_9)
```

```python
data_inf = pd.DataFrame(X_inference)
data_inf[['distance','name', 'surge_multiplier']]
```

# 4. Preparing Data to Predict 

```python
data_inf_cat= data_inf[list_cat_cols_nom]
data_inf_cat_ord= data_inf[list_cat_cols_ord]
data_inf_num = data_inf[list_num_cols]

data_inf_num_scaled= scaler.transform(data_inf_num)
data_inf_cat_ohe = ohe.transform(data_inf_cat).toarray()
data_inf_cat_encoded = encoder.transform(data_inf_cat_ord)
```

```python
data_inf_cat_ohe
```

```python
test_1 = np.concatenate([data_inf_num_scaled, data_inf_cat_ohe ,data_inf_cat_encoded], axis = 1)
```

# 5. Model's Prediction

```python
test_1_prediction = model.predict(test_1)
print(f'trip\'s cost prediction: {test_1_prediction}')
```
